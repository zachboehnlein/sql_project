These two  views are very useful and I used either similar views or Ctes to create for the other stocks in my database. This was particularly useful because I had to reference them quite a bit for data analysis. This was used so I could bring this into excel and compare them with other tables in my database. I reference these tables when joining data. In excel I was able to use these to come correlations and graph particular stocks to compare. This is written to help me compare stocks by a yearly basis. The justification for this is shown greater in my excel documents when I use these particular views to access data more quickly in order to get yearly comparisons of performance. The way we these was by using YEAR(date), AVG(close) and AVG(volume) in order get an accurate average per year for stock volume and close and to keep it in order we grouped by the date.

```python
%%sql 
CREATE OR REPLACE VIEW yearly_SPY_data AS 
    SELECT Year(SPY_date) AS date_by_year, Avg(SPY_close) AS close_by_year, AVG(SPY_volume) AS volume_by_year
    FROM SPY 
    GROUP BY Year(SPY_date);
```

```python
%%sql
CREATE OR REPLACE VIEW yearly_AAPL_data AS 
    SELECT Year(AAPL_date) AS date_by_year, Avg(AAPL_close) AS close_by_year, AVG(AAPL_volume) AS volume_by_year
    FROM AAPL
    GROUP BY Year(AAPL_date);
    
SELECT*
FROM yearly_AAPL_data 
```

This was used give me a metric for when a TSLA should be considered low or high on a recent historical basis. I used the last six months in order to at least find a recent metric for their stocks performace and set up a CASE to basically tell me when the stock is high, stable, low and when it is time to panic. This was done by looking at the stocks performance over the last six months and finding the highs and lows in order to see what my personal metric was for prices on the lower or higher end for TSLA would be. The justification for this would be to expand upon it and create something to autosell once it hits a certain low point or high point in order to either protect your investment or maximize it. This cases sql creates a new column that alerts based on TSLA_close price, the between was to have a recent comparison and we did not want it order by price but by date so we could see the most recent values. 

```python
%%sql
SELECT TSLA_date, TSLA_close, 
    CASE 
        WHEN TSLA_close < '550' THEN 'Panic'
        WHEN TSLA_close < '650' THEN 'Low' 
        WHEN TSLA_close BETWEEN '650' AND '750' THEN 'Stable'
        WHEN TSLA_close > '750'THEN 'High'
    END AS alerts 
FROM TSLA
WHERE TSLA_date BETWEEN '2019-12-01 00:00:00' AND '2020-05-01 00:00:00' 
ORDER BY TSLA_date;
```

Subquery to join the SPY table and yearly_wheat in order compare bushels to volume of stock traded withint the specific range of data available. This is justified as it is usefule for me to show how correlations can be relatively unhelpful when looking at data. And this will help me further explain why stock comparisons are tricky and should be taken will caution. This subquery took combined the tables on the year column again but we had to add a where statement on the SPY function as the bushels data only goes to 2014, so we didnt want any prices beyound that. Lasly we sorted it in Ascending order to be easier to read. 

```python
%%sql 
SELECT yearly_SPY_data.date_by_year, yearly_SPY_data.volume_by_year,
    (SELECT yearly_wheat.Bushels 
    FROM yearly_wheat
    WHERE yearly_SPY_data.date_by_year = yearly_wheat.Year
    ) AS bushels_by_year
FROM yearly_SPY_data
WHERE yearly_SPY_data.date_by_year < '2014'
ORDER BY yearly_SPY_data.date_by_year ASC;
```

This join was to use the views to compare yearly AAPL and SPY stock data in order to make conclusions about stock performance of AAPL when compared to SPY and to see how correlated to two stocks are which can be useful for futre buying or selling decisions. I joined together the two views from eariler on the date column which is the same and grouped by the SPY date as it has been around a lot less long so that it will exclude any AAPL closes before SPY was even a thing. 

```python
%%sql
SELECT yearly_SPY_data.date_by_year, yearly_SPY_data.close_by_year AS SPY_close, yearly_AAPL_data.close_by_year AS AAPL_close 
FROM yearly_SPY_data 
JOIN yearly_AAPL_data 
    ON yearly_SPY_data.date_by_year = yearly_AAPL_data.date_by_year
GROUP BY yearly_SPY_data.date_by_year;
```

This is very similar to a query above but done as a cte. This basically gets the data into a yearly format so I can work with it more easily by comparing stocks over the years instead of on a month to month basis. This time though we used CTE so this will not create a table for the TSLA stock. Personally I think the view works better generally. But this gives me a quick snapshot that I could call back to without creating the full table. The sytax for the code is extremely similar to the views above. 

```python
%%sql
WITH TSLA_yearly_cte AS (
    SELECT Year(TSLA_date) AS date_by_year, 
    Avg(TSLA_close) AS close_by_year, 
    AVG(TSLA_volume) AS volume_by_year
    FROM TSLA 
    GROUP BY date_by_year
) 
SELECT* 
FROM TSLA_yearly_cte 
```

The two windowed functions above were for stock comparision. In the background I had to set the close column to a decimal with 0 places in order to get good averages for volume. I took every close for TSLA and SPY and average out the volume on for closes with the same price. For example just above we see TSLA close and 19 underneather. This takes all the times the stock was at rounded to 19 and averages out the given volume. This gives us a metric of volatility and we can see how volume effects stock close price. This code was written similar to the view but taking the avg of volume as the stock prices withh be brought together due to the group by statement. 

```python
%%sql
SELECT 
    SPY_close,
    AVG(SPY_volume) OVER(PARTITION BY SPY_close) AS avg_volume
FROM SPY
WHERE SPY_date BETWEEN '2010-06-01' AND '2020-05-01'
GROUP BY SPY_close;
```

```python
%%sql
SELECT 
    TSLA_close,
    AVG(TSLA_volume) OVER(PARTITION BY TSLA_close) AS avg_volume
FROM TSLA 
WHERE TSLA_date BETWEEN '2010-06-01' AND '2020-05-01'
GROUP BY TSLA_close;
```

```python
%%sql
DELIMITER $$
DROP PROCEDURE IF EXISTS get_SPY_close_on_date;
CREATE PROCEDURE get_AAPL_close_on_date(in_AAPL_date DATETIME)
BEGIN
  SET @AAPL_date:=in_AAPL_date;
SELECT AAPL_date, AAPL_close
FROM AAPL 
WHERE AAPL_date = @AAPL_date
GROUP BY AAPL_date;
END$$
DELIMITER ;

CALL get_SPY_close_on_date('2020-01-01');
```
